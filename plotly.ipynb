```python
import pandas as pd
import numpy as np

import plotly
import plotly.express as px
import plotly.graph_objects as go
```

```python
# googleplaystore_1.csv is cleaned data downloaded from previous homework
df  = pd.read_csv("googleplaystore_1.csv")
```

```python
df
```

```python
avg_rating_per_category = df.groupby('Category')['Rating'].mean().reset_index()

fig = px.bar(avg_rating_per_category,
             y='Rating',
             x='Category',
             title='Average Rating by Category')
fig.show()
```

```python
install_counts = df['Installs'].value_counts()
labels = install_counts.index
values = install_counts.values
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.7)])
fig.update_layout(title='Distribution of Installs')
fig.show()
```

```python
data = go.Histogram(x=df['Rating'])

layout = dict(title="Histogram")

fig = go.Figure(data, layout)
fig.show()
```

```python
import plotly.express as px

non_free_apps = df[df['Type'] != 'Free']
fig = px.scatter(non_free_apps, y='Rating', x='Price, $', title='Scatterplot of Rating vs Price for Non-Free Apps')
fig.show()
```

```python
fig = px.box(df, x="Type", y="Rating", title="Boxplot of Rating by Type")
fig.show()
```
